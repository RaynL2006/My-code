```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def distance_finder(x, y):
    x2 = x**2
    y2 = y**2
    distance = math.sqrt(x2 + y2)
    return distance

def Question_1(num_points):
    num_points_in_circle = 0

    for i in range(num_points):
        # Pick a point from a square with centre (0,0) and side length 1
        x_ord = random.uniform(-0.5, 0.5)
        y_ord = random.uniform(-0.5, 0.5)

        if distance_finder(x_ord, y_ord) < 0.5: # ie. if the point lies in the circle with diameter 1 centered at (0, 0)
            num_points_in_circle += 1

    pi_estimate = 4 * num_points_in_circle / num_points

    return pi_estimate

Question_1(10000000)
```

```python
def Question_2(num_runs):
    run_displacements = []

    for run in range(num_runs):
        total_run_displacement = 0

        for step in range(100):
            step_displacement = random.choice([-1, 1])
            total_run_displacement += step_displacement
        
        run_displacements.append(total_run_displacement)
    
    mean_run_displacement = sum(run_displacements) / num_runs

    return mean_run_displacement

Question_2(10000)
```

```python
def Question_3(num_samples, X_lower_bound, X_upper_bound):

    list_of_X_bars = []
    list_of_Xs = []

    for sample in range(num_samples):
        four_values = []

        for i in range(4):
            X = random.uniform(X_lower_bound, X_upper_bound)
            four_values.append(X)
            list_of_Xs.append(X)
        
        X_bar = sum(four_values) / 4

        list_of_X_bars.append(X_bar)
    
    
    plt.hist(np.array(list_of_Xs))
    plt.xlabel('X')
    plt.ylabel('Frequency')
    plt.title('Basic histogram (frequency - not frequency density) vs X')
    plt.show()

    plt.hist(np.array(list_of_X_bars))
    plt.xlabel('X bar')
    plt.ylabel('Frequency')
    plt.title('Basic histogram (frequency - not frequency density) vs X bar')
    plt.show()

Question_3(10000, -10, 10)
```
